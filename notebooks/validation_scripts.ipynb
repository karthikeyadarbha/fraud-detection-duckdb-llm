```python
import pandas as pd, numpy as np
from sklearn.metrics import roc_auc_score, average_precision_score

res = pd.read_parquet('artifacts/results.parquet')
orig = pd.read_csv('/tmp/sample_10k.csv', usecols=['transaction_id','is_fraud'])
m = orig.merge(res, on='transaction_id', how='left')

y = m['is_fraud'].astype(str).str.upper().map({'TRUE':1,'FALSE':0}).fillna(0).astype(int)
print("rows:", len(m), "positives:", y.sum())
print("Baseline AUC:", roc_auc_score(y, m['baseline_score']))
print("Baseline PR-AUC:", average_precision_score(y, m['baseline_score']))
print("Combined (final_score) AUC:", roc_auc_score(y, m['final_score']))
print("Combined PR-AUC:", average_precision_score(y, m['final_score']))

# precision@k for Baseline and Final
for k in (50,100,200):
    topb = m.sort_values('baseline_score', ascending=False).head(k)['is_fraud'].astype(str).str.upper().map({'TRUE':1,'FALSE':0}).sum()
    topf = m.sort_values('final_score', ascending=False).head(k)['is_fraud'].astype(str).str.upper().map({'TRUE':1,'FALSE':0}).sum()
    print(f"Precision@{k}: baseline={topb}/{k}={(topb/k):.3f}, final={topf}/{k}={(topf/k):.3f}")
```

```python
%%bash

#!/usr/bin/env bash
set -euo pipefail

REPO_ROOT="/workspaces/fraud-detection-duckdb-llm"
BACKUP_DIR="/tmp/backup_fraud_project_$(date +%Y%m%d%H%M%S)"
MIN_SIZE="50M"   # threshold for "large" files to move; adjust if you want different cutoff

echo "Repo root: $REPO_ROOT"
echo "Backup directory: $BACKUP_DIR"
echo "Finding files >= $MIN_SIZE in repo..."

mkdir -p "$BACKUP_DIR"

# List largest directories in repo (top 20)
echo
echo "==== Top directories in repo ===="
du -sh "$REPO_ROOT"/* 2>/dev/null | sort -h | tail -n 20 || true

echo
echo "==== Large files (>= $MIN_SIZE) found (preview) ===="
find "$REPO_ROOT" -type f -size +$MIN_SIZE -exec ls -lh {} \; 2>/dev/null | awk '{print $5 " " $9}' | sort -h || true

read -p "Move these large files to $BACKUP_DIR to free workspace space? (y/N) " CONF
if [[ "${CONF:-N}" != "y" && "${CONF:-Y}" != "Y" && "${CONF:-Y}" != "y" ]]; then
  echo "Aborting move of large files. You can still run other cleanup steps below manually."
else
  echo "Moving files..."
  find "$REPO_ROOT" -type f -size +$MIN_SIZE -print0 2>/dev/null | while IFS= read -r -d '' file; do
    dest="$BACKUP_DIR/$(dirname "$file" | sed "s|$REPO_ROOT/||")"
    mkdir -p "$dest"
    echo "mv '$file' -> '$dest/'"
    mv -v "$file" "$dest/" || echo "FAILED to move $file"
  done
  echo "Moved large files to $BACKUP_DIR"
fi

echo
read -p "Remove pip cache (~/.cache/pip)? (y/N) " CONF
if [[ "${CONF:-N}" == "y" || "${CONF:-Y}" == "Y" ]]; then
  rm -rf "$HOME/.cache/pip"
  echo "Removed pip cache"
fi

read -p "Remove user caches (~/.cache/*)? (may remove non-critical caches) (y/N) " CONF
if [[ "${CONF:-N}" == "y" || "${CONF:-Y}" == "Y" ]]; then
  rm -rf "$HOME/.cache"/* || true
  echo "Removed user caches"
fi

read -p "Remove Python venvs named .venv or venv in repo? (y/N) " CONF
if [[ "${CONF:-N}" == "y" || "${CONF:-Y}" == "Y" ]]; then
  rm -rf "$REPO_ROOT/.venv" "$REPO_ROOT/venv" || true
  echo "Removed virtualenvs (.venv, venv) in repo if present"
fi

read -p "Remove Jupyter checkpoints and __pycache__ directories in the repo? (y/N) " CONF
if [[ "${CONF:-N}" == "y" || "${CONF:-Y}" == "Y" ]]; then
  find "$REPO_ROOT" -name ".ipynb_checkpoints" -type d -exec rm -rf {} + || true
  find "$REPO_ROOT" -type d -name "__pycache__" -exec rm -rf {} + || true
  find "$REPO_ROOT" -name '*.pyc' -delete || true
  echo "Removed notebook checkpoints and python caches"
fi

read -p "Run 'sudo apt-get clean' to free apt cache? (may require sudo) (y/N) " CONF
if [[ "${CONF:-N}" == "y" || "${CONF:-Y}" == "Y" ]]; then
  sudo apt-get clean || true
  echo "apt cache cleaned"
fi

# Docker prune (optional and destructive)
if command -v docker >/dev/null 2>&1; then
  read -p "Run 'sudo docker system prune -af --volumes' to remove docker images/containers/volumes? (very destructive) (y/N) " CONF
  if [[ "${CONF:-N}" == "y" || "${CONF:-Y}" == "Y" ]]; then
    sudo docker system prune -af --volumes || true
    echo "Docker system prune done"
  fi
fi

# Extremely destructive: remove all untracked files in repo
read -p "RUN 'git clean -fdx' in $REPO_ROOT (removes ALL untracked/ignored files)? This is DESTRUCTIVE. (y/N) " CONF
if [[ "${CONF:-N}" == "y" || "${CONF:-Y}" == "Y" ]]; then
  (cd "$REPO_ROOT" && git clean -fdx) || true
  echo "git clean completed in repo"
fi

# Git housekeeping to free local repo objects
echo "Running git housekeeping..."
(cd "$REPO_ROOT" && git reflog expire --expire=now --all 2>/dev/null || true)
(cd "$REPO_ROOT" && git gc --prune=now --aggressive 2>/dev/null || true)

echo "Final disk usage:"
df -h .

echo
echo "DONE. Large files (if moved) are in: $BACKUP_DIR"
echo "If you still see no freed space, try restarting the Codespace in the GitHub UI (some overlay reclaiming happens on restart)."
```

```python
%%bash

#!/usr/bin/env bash
# Interactive Conda cleanup helper for Codespaces
# Inspect before running. Prompts before each destructive action.

set -euo pipefail

echo "=== Conda cleanup helper ==="
echo

# Where conda root is usually installed in Codespaces
CONDA_ROOT="/opt/conda"
PKG_CACHE="$CONDA_ROOT/pkgs"
USER_CONDA_DIR="$HOME/.conda"

echo "Conda root (expected): $CONDA_ROOT"
echo "Conda pkg cache: $PKG_CACHE"
echo "User conda dir: $USER_CONDA_DIR"
echo

# Show basic info
echo "conda info (short):"
conda info --json 2>/dev/null || conda info || true
echo

echo "Disk usage summary:"
df -h .
du -sh "$CONDA_ROOT" 2>/dev/null || true
du -sh "$PKG_CACHE" 2>/dev/null || true
du -sh "$USER_CONDA_DIR" 2>/dev/null || true
echo

read -p "Run conda clean --all to remove caches and tarballs (recommended)? (y/N) " ans
if [[ "${ans:-N}" == "y" || "${ans:-Y}" == "Y" ]]; then
  echo "Running: conda clean --all -y"
  conda clean --all -y || echo "conda clean returned an error"
fi

# List environments and sizes
echo
echo "Conda environments:"
conda env list
echo
echo "Sizes of env directories under $CONDA_ROOT/envs (if present):"
du -sh "$CONDA_ROOT/envs"/* 2>/dev/null | sort -h || true
du -sh "$USER_CONDA_DIR/envs"/* 2>/dev/null | sort -h || true

# Offer to remove envs interactively
echo
read -p "Would you like to interactively remove specific conda envs? (y/N) " ans
if [[ "${ans:-N}" == "y" || "${ans:-Y}" == "Y" ]]; then
  echo "Listing envs:"
  conda env list
  echo
  while true; do
    read -p "Enter environment name to REMOVE (or press Enter to finish): " envname
    if [[ -z "${envname}" ]]; then
      break
    fi
    read -p "Confirm removal of conda env '$envname' (this will delete the env)? (type 'REMOVE' to confirm) " confirm
    if [[ "${confirm}" == "REMOVE" ]]; then
      echo "Removing env: $envname"
      conda env remove -n "$envname" -y || echo "Failed to remove env $envname"
      conda clean --all -y || true
    else
      echo "Skipped $envname"
    fi
  done
fi

# Package cache details & manual prune
echo
du -sh "$PKG_CACHE"/* 2>/dev/null | sort -h | tail -n 20 || true
read -p "Remove large package cache files under $PKG_CACHE (tarballs/conda files)? (y/N) " ans
if [[ "${ans:-N}" == "y" || "${ans:-Y}" == "Y" ]]; then
  echo "Removing tarballs and cache files under $PKG_CACHE (non-recoverable)"
  find "$PKG_CACHE" -type f -name '*.tar.bz2' -delete || true
  find "$PKG_CACHE" -type f -name '*.conda' -delete || true
  # optionally remove cache dirs containing package caches:
  rm -rf "$PKG_CACHE"/*cache* 2>/dev/null || true
  conda clean --all -y || true
fi

# Pip and user caches
echo
read -p "Remove pip cache (~/.cache/pip)? (y/N) " ans
if [[ "${ans:-N}" == "y" || "${ans:-Y}" == "Y" ]]; then
  rm -rf "$HOME/.cache/pip" || true
  echo "Removed pip cache"
fi

read -p "Remove user caches (~/.cache/*)? (may affect non-critical caches) (y/N) " ans
if [[ "${ans:-N}" == "y" || "${ans:-Y}" == "Y" ]]; then
  rm -rf "$HOME/.cache"/* || true
  echo "Removed user caches"
fi

# Final housekeeping
echo
echo "Running final conda clean and git housekeeping..."
conda clean --all -y || true
git -C /workspaces/fraud-detection-duckdb-llm reflog expire --expire=now --all 2>/dev/null || true
git -C /workspaces/fraud-detection-duckdb-llm gc --prune=now --aggressive 2>/dev/null || true

echo
echo "Disk usage after cleanup:"
df -h .
echo
echo "NOTE: If overlay usage still shows 100% (or no space freed), restart the Codespace from GitHub UI."
echo "Backups (if any) are left in /tmp by earlier scripts - move them to persistent storage if needed."
echo "Done."
```
